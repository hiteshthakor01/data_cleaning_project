```python
# importing libraries
import numpy as np
import pandas as pd
```

```python
# reading data
df = pd.read_csv('/workspaces/data_cleaning_project/data/Cafe Sales.csv')
```

```python
# inspecting first few records
df.head()
```

```python
# inspecting last few records
df.tail()
```

```python
# data structure
df.info()
```

```python
# duplicate records
sum(df.duplicated())
```

```python
# missing records
df.isna().sum() * 100 / df.shape[0]
```

```python
# unique records
df_columns = set(df.columns)
exclude = {'Transaction ID'}
df_columns = df_columns.difference(exclude)
```

```python
for column in df_columns:
    print(f'{column}:\n{df[column].unique()}', end = '\n\n')
```

```python
# .replace()
df = df.replace(['ERROR', 'UNKNOWN'], np.nan)
```

```python
# inspecting first few records
df.head()
```

```python
# missing records
df.isna().sum() * 100 / df.shape[0]
```

```python
# fixing data types
df['Quantity'] = df['Quantity'].astype('Int64')             # when replacing .nan with int, always use Int64
df[['Price Per Unit', 'Total Spent']] = df[['Price Per Unit', 'Total Spent']].astype('float')
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'])
```

```python
# dropping records where Transaction Date is missing
df = df.dropna(subset = ['Transaction Date'])
```

```python
# replacing missing values in Payment Method and Location
df[['Payment Method', 'Location']] = df[['Payment Method', 'Location']].fillna('INA')
```

```python
# replacing missing values in Item with mode value
df['Item'] = df['Item'].fillna(df['Item'].mode()[0])
```

```python
# replacing missing values in Quantity with median value
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].median())
```

```python
# inspecting Price Per Unit by Item
df.groupby('Item')['Price Per Unit'].value_counts()
```

```python
# filling missing values in Price Per Unit with mode value of Item category
df['Price Per Unit'] = df.groupby('Item')['Price Per Unit'].transform(lambda x: x.fillna(x.mode()[0]))
```

```python
# filling missing values in Total Spent
df['Total Spent'] = df['Total Spent'].fillna(df['Quantity'] * df['Price Per Unit'])
```

```python
# saving cleansed data
df.to_csv('/workspaces/data_cleaning_project/data/Cafe Sales_Cleaned.csv')
```
